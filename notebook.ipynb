# Project: Analysis of Wars Involving the United States (XX-XXI Century)

## Introduction
In this data-driven project, we aim to explore the historical trends of wars involving the United States during the 20th and 21st centuries. Our focus is on identifying patterns related to the political party affiliation of the president at the time a war started and ended.

## Project Goal
The primary goal is to examine whether there is a discernible pattern in the initiation and conclusion of wars by the two major political parties in the United States: Democrats and Republicans.

## Methodology

### Data Collection

Our data is sourced from the Wikipedia page on [List of wars involving the United States](https://en.wikipedia.org/wiki/List_of_wars_involving_the_United_States). We accessed this comprehensive list, extracting relevant information about wars in the XX and XXI centuries.

**Data Source:**
The war data used in this analysis was sourced from [Wikipedia - List of wars involving the United States](https://en.wikipedia.org/wiki/List_of_wars_involving_the_United_States).

### Data Translation and Analysis

The collected data was processed and translated into a structured format suitable for analysis. We focused on key attributes such as the starting and ending years of each war, the president's party affiliation at the war's inception and end, and whether the war is ongoing.

### Question and Analysis

The primary question guiding our analysis was whether there is a notable difference in the likelihood of a war starting under a Democratic or Republican president. We approached this question by examining the frequency of wars initiated by each political party.

## Statistical Analysis

In addition to the visual representation of wars started and ended by different political parties, let's perform a basic statistical analysis to calculate the probabilities.

### Probability of a US President Starting a War

Let "P(D)" represent the probability that a US president will start a war if they are a Democrat, and "P(R)" represent the probability that a US president will start a war if they are a Republican.

    P(D) = (Number of wars started by Democrats) / (Total number of wars)
    P(R) = (Number of wars started by Republicans) / (Total number of wars)

Now, let's calculate these probabilities based on our dataset.

```python
# Calculate probabilities
total_wars = df.shape[0]
democrat_started = df[df['Party at War Start'] == 'Democrat'].shape[0]
republican_started = df[df['Party at War Start'] == 'Republican'].shape[0]

# Probability calculations
probability_democrat = democrat_started / total_wars
probability_republican = republican_started / total_wars

print("Probability of a US president starting a war if Democrat:", probability_democrat)
print("Probability of a US president starting a war if Republican:", probability_republican)
```

```python
# Data for the pie chart
probabilities = [probability_democrat, probability_republican]
labels = ['Democrat', 'Republican']
colors = ['blue', 'red']

# Create a pie chart
plt.figure(figsize=(6, 6))
plt.pie(probabilities, labels=labels, colors=colors, autopct='%1.1f%%', startangle=140)
plt.title('Probability of a US President Starting a War by Political Party')
plt.show()
```
