```python
import matplotlib.pyplot as plt
import seaborn as sns

# Convert "present" to NaN for plotting
df['Year End'] = pd.to_numeric(df['Year End'], errors='coerce')

# Map party colors
party_colors = {'Republican': 'red', 'Democrat': 'blue'}

# Select relevant columns
president_info = df[['War', 'Party at War Start', 'Year Start', 'Year End', 'Party at War End']]

# Create a scatter plot
plt.figure(figsize=(12, 8))
sns.scatterplot(x='Year Start', y='War', hue='Party at War Start', data=president_info, s=100, palette=party_colors, legend='full')
sns.scatterplot(x='Year End', y='War', hue='Party at War End', data=president_info[president_info['Year End'].notnull()], s=100, palette=party_colors, legend='full', marker='o')

# Set labels and title
plt.xlabel('Year')
plt.ylabel('War')
plt.title('Political Party Affiliation at the Start and End of Each War')
plt.legend(title='Party at War Start', bbox_to_anchor=(1.05, 1), loc='upper left')

# Show the plot
plt.show()
```

```python
# Replace NaN values with 'present' in 'Year End' column
df['Year End'] = df['Year End'].replace(np.nan, 'present')

# Wars started by a Democrat president
democrat_started = df[df['Party at War Start'] == 'Democrat'].shape[0]

# Wars started by a Republican president
republican_started = df[df['Party at War Start'] == 'Republican'].shape[0]

# Wars ended by a Democrat president
democrat_ended = df[df['Party at War End'] == 'Democrat'].shape[0]

# Wars ended by a Republican president
republican_ended = df[df['Party at War End'] == 'Republican'].shape[0]

# Wars started and ended by Democrats
democrat_started_ended = df[(df['Party at War Start'] == 'Democrat') & (df['Party at War End'] == 'Democrat')].shape[0]

# Wars started and ended by Republicans
republican_started_ended = df[(df['Party at War Start'] == 'Republican') & (df['Party at War End'] == 'Republican')].shape[0]

# Wars started by Democrats and ended by Republicans
democrat_started_ended_republican = df[(df['Party at War Start'] == 'Democrat') & (df['Party at War End'] == 'Republican')].shape[0]

# Wars started by Republicans and ended by Democrats
republican_started_ended_democrat = df[(df['Party at War Start'] == 'Republican') & (df['Party at War End'] == 'Democrat')].shape[0]

# Ongoing wars started by Democrats
ongoing_democrat = df[(df['Year End'].str.lower().str.contains('present')) & (df['Party at War Start'] == 'Democrat')].shape[0]

# Ongoing wars started by Republicans
ongoing_republican = df[(df['Year End'].str.lower().str.contains('present')) & (df['Party at War Start'] == 'Republican')].shape[0]

# Display the results
print(f'Wars started by a Democrat president: {democrat_started}')
print(f'Wars started by a Republican president: {republican_started}')
print(f'Wars ended by a Democrat president: {democrat_ended}')
print(f'Wars ended by a Republican president: {republican_ended}')
print(f'Wars started and ended by Democrats: {democrat_started_ended}')
print(f'Wars started and ended by Republicans: {republican_started_ended}')
print(f'Wars started by Democrats and ended by Republicans: {democrat_started_ended_republican}')
print(f'Wars started by Republicans and ended by Democrats: {republican_started_ended_democrat}')
print(f'Ongoing wars started by Democrats: {ongoing_democrat}')
print(f'Ongoing wars started by Republicans: {ongoing_republican}')
```
